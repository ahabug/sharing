```python
def check_equal(item1, item2): # 1 true 0 false
    if np.isnan(item1) and np.isnan(item2):
        return 1
    elif np.isnan(item1) or np.isnan(item2):
        return 0
    elif (item1==0) and (item2==0):
        return 1
    elif (item1==0) or (item2==0):
        return 0
    elif np.abs(np.log(item1/item2)) <0.1:
        return 1
    else:
        return 0
```
